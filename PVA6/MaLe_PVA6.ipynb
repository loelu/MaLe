## 1.1 sklearn.linear_model.LinearRegression

```python
diabetes_dataset_X, diabetes_dataset_y = datasets.load_diabetes(return_X_y=True)

# use only one feature to show a 2d graph of the result
diabetes_dataset_X = diabetes_dataset_X[:, np.newaxis, 2]

# split into train and test data
diabetes_dataset_X_train = diabetes_dataset_X[:-20]
diabetes_dataset_X_test = diabetes_dataset_X[-20:]

diabetes_dataset_y_train = diabetes_dataset_y[:-20]
diabetes_dataset_y_test = diabetes_dataset_y[-20:]

# linear regression
regr = linear_model.LinearRegression()

# model training
regr.fit(diabetes_dataset_X_train, diabetes_dataset_y_train)

# predictions
diabetes_dataset_y_predictions = regr.predict(diabetes_dataset_X_test)


print("Performance of the model:")

# The root mean squared error: 0 is perfect prediction
print('Root Mean squared error: %.2f'
      % math.sqrt(mean_squared_error(diabetes_dataset_y_test, diabetes_dataset_y_predictions)))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(diabetes_dataset_y_test, diabetes_dataset_y_predictions))

# Plot outputs
plt.scatter(diabetes_dataset_X_test, diabetes_dataset_y_test,  color='black')
plt.plot(diabetes_dataset_X_test, diabetes_dataset_y_predictions, color='blue', linewidth=3)
plt.show()
```

### 3.1 Diabetes Dataset

- R2 Score (Coefficient of determination): 0.47 (lineare Regression) und -0.23 (DecisionTreeRegressor). Der R2 Score f√ºr linearen Regression ist naher bei 1, und deshalb besser. Der Score vom DecisionTreeRegressor ist sogar negativ.. (Reference: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html)

Dieses Ergebnis wurde berechnet, wenn bei X nur 1 Feature verwendet wurde. Im einem n√§chsten Versuch wurden alle Features (10) beim Training verwendet und beide Modelle wurden sichtlich besser, jedoch bleibt, das Lineare-Regressions-Modell besser als das DecisionTreeRegressor-Modell.

Beim Boston Dataset zeigt sich dasselbe Bild, wie beim Diabetes Dataset:
