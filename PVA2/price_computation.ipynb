```python
# param: normalized_base_price, float. A value between 0 and 1
def value_color(normalized_base_price: float):
    if normalized_base_price < 0.3:
        return "green"
    elif normalized_base_price >= 0.3 and normalized_base_price < 0.6:
        return "orange"
    else:
        return "red"

# param: product, JSON from Migros API /products/:id
# param: product_list, an array of products
def price_comparison(product, product_list):
    all_prices = list(map(lambda x: x['price']['base']['price'], product_list))

    max_price = float(max(all_prices))
    min_price = float(min(all_prices))
    base_price = product['price']['base']['price']
    
    # receive base_price between 0 and 1
    normalized_base_price = (base_price-min_price)/(max_price-min_price)
    
    index_max_price = all_prices.index(max_price)
    index_min_price = all_prices.index(min_price)
    
    price_data = {
        "item_price": product['price']['item'],
        "base_price": product['price']['base'],
        "max_price": product_list[index_max_price]['price']['base'],
        "min_price": product_list[index_min_price]['price']['base'],
        "normalized_base_price": normalized_base_price,
        "base_price_color": value_color(normalized_base_price)
    }
    
    return price_data
```
